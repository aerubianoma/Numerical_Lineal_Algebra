```julia
using LinearAlgebra

```

```julia
x = rand(5,1)
y = rand(5,1)
xy = x'y
```

```julia
p=1+exp(randn())
q=1/(1-1/p)
println("p=",p,"\t q=",q)
println("||y||_p*||x||_q = ", norm(y,p)*norm(x,q))
println("||yx||_2 = ",xy[1])
```

```julia
p = 1 + exp(randn())
q = p + exp(randn())
n = 300 
x = rand(n)

#println(x')
println(norm(x,q))
println(norm(x,p))
println((n^(1/p-1/q)*norm(x,q)))
```
